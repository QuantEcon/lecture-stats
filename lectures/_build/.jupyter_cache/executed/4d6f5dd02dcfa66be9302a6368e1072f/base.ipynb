```python
import numpy as np
from numba import njit, vectorize, prange
import matplotlib.pyplot as plt
from math import gamma
```

```python
# Parameters in the two beta distributions.
F_a, F_b = 1, 1
G_a, G_b = 3, 1.2

@vectorize
def p(w, a, b):
    r = gamma(a + b) / (gamma(a) * gamma(b))
    return r * w ** (a-1) * (1 - w) ** (b-1)

# The two density functions.
f = njit(lambda w: p(w, F_a, F_b))
g = njit(lambda w: p(w, G_a, G_b))
```

```python
w_range = np.linspace(1e-5, 1-1e-5, 1000)

plt.plot(w_range, g(w_range), label='g')
plt.plot(w_range, f(w_range), label='f')
plt.xlabel('$\omega$')
plt.legend()
plt.title('density functions $f$ and $g$')
plt.show()
```

```python
l = njit(lambda w: f(w) / g(w))
```

```python
plt.plot(w_range, l(w_range))
plt.title('$\ell(\omega)$')
plt.xlabel('$\omega$')
plt.show()
```

```python
g_a, g_b = G_a, G_b
h_a, h_b = 0.5, 0.5
```

```python
w_range = np.linspace(1e-5, 1-1e-5, 1000)

plt.plot(w_range, g(w_range), label=f'g=Beta({g_a}, {g_b})')
plt.plot(w_range, p(w_range, 0.5, 0.5), label=f'h=Beta({h_a}, {h_b})')
plt.title('real data generating process $g$ and importance distribution $h$')
plt.legend()
plt.ylim([0., 3.])
plt.show()
```

```python
@njit(parallel=True)
def estimate(p_a, p_b, q_a, q_b, T=1, N=10000):

    _L = 0
    for i in prange(N):

        L = 1
        weight = 1
        for t in range(T):
            w = np.random.beta(q_a, q_b)
            l = f(w) / g(w)

            L *= l
            weight *= p(w, p_a, p_b) / p(w, q_a, q_b)

        _L += L * weight

    _L /= N

    return _L
```

```python
estimate(g_a, g_b, g_a, g_b, T=1, N=10000)
```

```python
estimate(g_a, g_b, h_a, h_b, T=1, N=10000)
```

```python
estimate(g_a, g_b, g_a, g_b, T=10, N=10000)
```

```python
estimate(g_a, g_b, h_a, h_b, T=10, N=10000)
```

```python
@njit(parallel=True)
def simulate(p_a, p_b, q_a, q_b, N_simu, T=1):

    _L_p = np.empty(N_simu)
    _L_q = np.empty(N_simu)

    for i in prange(N_simu):
        _L_p[i] = estimate(p_a, p_b, p_a, p_b, T=T)
        _L_q[i] = estimate(p_a, p_b, q_a, q_b, T=T)

    return _L_p, _L_q
```

```python
N_simu = 1000
_L_p, _L_q = simulate(g_a, g_b, h_a, h_b, N_simu)
```

```python
# standard Monte Carlo (mean and std)
np.nanmean(_L_p), np.nanvar(_L_p)
```

```python
# importance sampling (mean and std)
np.nanmean(_L_q), np.nanvar(_L_q)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

_range = np.linspace(0, 2, 100)

for i, t in enumerate([1, 5, 10, 20]):
    row = i // 2
    col = i % 2

    _L_p, _L_q = simulate(g_a, g_b, h_a, h_b, N_simu, T=t)
    _hat_p, _hat_q = np.nanmean(_L_p), np.nanmean(_L_q)
    _hat_p, _hat_q = np.nanvar(_L_p), np.nanvar(_L_q)

    axs[row, col].set_xlabel('$_L$')
    axs[row, col].set_ylabel('frequency')
    axs[row, col].set_title(f'$T$={t}')
    n_p, bins_p, _ = axs[row, col].hist(_L_p, bins=_range, color='r', alpha=0.5, label='$g$ generating')
    n_q, bins_q, _ = axs[row, col].hist(_L_q, bins=_range, color='b', alpha=0.5, label='$h$ generating')
    axs[row, col].legend(loc=4)

    for n, bins, _hat, _hat in [[n_p, bins_p, _hat_p, _hat_p],
                                  [n_q, bins_q, _hat_q, _hat_q]]:
        idx = np.argmax(n)
        axs[row, col].text(bins[idx], n[idx], '$\hat{}$='+f'{_hat:.4g}'+', $\hat{}=$'+f'{_hat:.4g}')

plt.show()
```
